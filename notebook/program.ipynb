PROBLEM

Terdapat data dari aplikasi pesan antar makanan berupa food delivery dataset. Dari dataset tersebut akan dicari frekuensi pembelian setiap hari dan setiap jamnya

ALGORITMA
1. Import dataset,
2. gunakan kolom resto_id, date_time, dan status
3.cleaning data yang status=complete

1. import library dan dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
PATH_DATA='../dataset/'
```

```python
df_trx = pd.read_csv(PATH_DATA+'food_delivery_datasets.csv')
```

```python
df_trx 
```

```python
t1 = df_trx[['cust_id','status']]
```

```python
t1
```

```python
df_trx['date_time'] = pd.to_datetime(df_trx['date_time']).dt.round('H')
```

```python
df_trx['date_time']
```

```python
t2= df_trx['date_time'].dt.strftime('%A')
```

```python
t2
```

```python
t3= df_trx['date_time'].dt.strftime('%H:%M')
```

```python
t3
```

```python
df_trx= pd.concat([t1['cust_id'], t2 , t3, t1['status']],axis=1)
df_trx.columns = ['customer_id', 'days', 'time', 'status']
```

```python
df_trx
```

2. Cleaning dataset dengan status complete

```python
group_status = df_trx.groupby('status')
group_status['status'].size()
```

```python
df_trx= df_trx[df_trx.status == 'Completed']
df_trx.reset_index(drop=True)
```

4. Visualisasi data

```python
# Data grouping based on days of the period.

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plot1 = df_trx.groupby(['days']).size().reindex(days).plot(kind='bar', figsize = (8,6), ylabel='frekuensi', title='Frekuensie perhari')
plot1.figure.savefig('../freq-over-days.jpg')
```

```python
df_trx.groupby(['days']).size()
```

Dari data tersebut terlihat bahwa order paling banyak terjadi pada minggu dan sabtu

```python
# Data grouping based on time of the period. (cumulative)

fig, ax = plt.subplots()
hours = mdates.HourLocator(interval = 1)

hm = df_trx.groupby(['time']).size()

ax.plot(hm.index, hm.values)
hm.plot(ax = ax, x_compat=True, figsize = (8,6), ylabel='frekuensi', 
title='Frekuensi perwaktu')
ax.xaxis.set_major_locator(hours)

fig.autofmt_xdate()
fig.savefig('../freq-over-cumtimes.jpg')
```

```python
df_trx.groupby(['time']).size()
```

Dari data tersebut terlihat bahwa order paling banyak terjadi pada jam makan siang dan makan malam yaitu sekitaran pukul 12.00-18.00

```python
# Data grouping based on time of the days of the period.

group_days = df_trx.groupby(['days', 'time']).size().unstack(level=0, fill_value=0)
plot3 = group_days.plot(figsize=(8,6), ylabel='frekuensi',
title='grafik hubungan frekuensi hari dan frekuensi waktu')
                
plot3.figure.savefig('../freq-over-times-multi.jpg')
```

Dari grafik terlihat bahwa orderan paling banyak terjadi pada hari minggu sekitar pukul 12.00-17.00

Dari data dan grafik terlampir, disimpulkan bahwa orderan paling banyak terjadi pada hari minggu sekitar pukul 14.00
